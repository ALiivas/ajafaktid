## 1. Andmete sisselugemine

JSON kujul andmete sisselugemiseks kasutatakse `estnltk.converters` moodulist funktsiooni `json_to_text`, mis loeb JSON failist sisse Text objekti ning tagastab selle.

```python
from estnltk.converters import json_to_text
import os

path = 'korpus_json/'
text_objects = []
for file in os.listdir(path):
    filename = os.path.join(path, file)
    text = json_to_text(file=filename)
    text_objects.append(text)
```

## 3. Seoseliikmete väljavõtmine

Seoseliikmed on EstNLTK-s alati nende lauses esinemise järjekorras ehk a_text on lauses alati enne kui b_text. BRAT-is võis järjekord olla ka teine. Seda tuleb võtta arvesse, kui on soov tuvastada näiteks ajaseosetüüpide esinemissagedusi korpuses: kui ajaseose esimene liige ei ole ajaväljend, vaid on entiteet/sündmus, siis tuleb seosetüüp pöörata ümber, sest BRAT-is on ajaseos tõmmatud alati ajaväljendist sündmuse/entiteedini. Argumentseoste puhul sündmuste vahel näitab liikmete vastupidist järjekorda lauses see, kui seosetüüp on *has_argument* asemel *is_argument_of*. Sellisel juhul on argumenti nõudev sündmus b_text ning a_text selle argument.

Seosed on EstNLTK-s justkui katkendlikud fraasid. Seose atribuut b_index ütleb, millisest indeksist alates algab seose fraasis teine sõna. Järgnevat funktsiooni saab kasutada kas ajaseoste või argumentseoste liikmete kättesaamiseks.

```python
#Funktsioon mõlema seoseliikme kättesaamiseks. Tagastab Annotation objektid, milles on märgenduse infosisu.
def seosed(link, text):
    esimene_el = link.spans[0:link.b_index[0]]
    teine_el = link.spans[link.b_index[0]:]
    esimene_liige = ""
    teine_liige = ""
    if text.timexes[ lambda x: x.brat_id == esimene_el[0].brat_id ]:
        esimene_liige = text.timexes[ lambda x: x.brat_id == esimene_el[0].brat_id ].spans[0].annotations[0]
        if text.timexes[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.timexes[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
        elif text.events[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.events[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
        elif text.entities[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.entities[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
    elif text.events[ lambda x: x.brat_id == esimene_el[0].brat_id ]:
        esimene_liige = text.events[ lambda x: x.brat_id == esimene_el[0].brat_id ].spans[0].annotations[0]
        if text.timexes[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.timexes[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
        elif text.events[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.events[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
        elif text.entities[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.entities[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
    elif text.entities[ lambda x: x.brat_id == esimene_el[0].brat_id ]:
        esimene_liige = text.entities[ lambda x: x.brat_id == esimene_el[0].brat_id ].spans[0].annotations[0]
        if text.timexes[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.timexes[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
        elif text.events[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.events[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
        elif text.entities[ lambda x: x.brat_id == teine_el[0].brat_id ]:
            teine_liige = text.entities[ lambda x: x.brat_id == teine_el[0].brat_id ].spans[0].annotations[0]
    return esimene_liige, teine_liige
```

```python
#Kasutusnäide esimese Text objekti ajaseoste peal
text = text_objects[0]
for i in range(len(text.tlinks)):
    link = text.tlinks[i]
    esimene_liige, teine_liige = seosed(link, text)
```

```python
#Kasutusnäide teise Text objekti argumentseoste peal
text = text_objects[1]
for i in range(len(text.event_arguments)):
    link = text.event_arguments[i]
    esimene_liige, teine_liige = seosed(link, text)
```
